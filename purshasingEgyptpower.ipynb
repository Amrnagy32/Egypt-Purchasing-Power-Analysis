```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

# Task 1.0: Inflation Analysis

```python
inflation_df = pd.read_csv("API_FP.CPI.TOTL.ZG_DS2_en_csv_v2_32.csv", skiprows = 4)
inflation_df.head()
```

# 1.1 Egypt Inflation

```python
inflation_egypt = inflation_df[inflation_df["Country Name"] == "Egypt, Arab Rep."]
inflation_egypt.head()
```

```python
years = [str(year) for year in range(2010, 2025)]
```

```python
inflation_sel = inflation_egypt[years]
```

```python
inflation_sel.head()
```

# 1.2 Reset Indexing 

```python
inflation_long = inflation_sel.T.reset_index()
inflation_long.columns = ["Year", "Inflation"]
inflation_long.head(15)
```

```python
inflation_long["Year"] = inflation_long["Year"].astype(int)
```

```python
inflation_long.info()
```

# 1.3 Ploting result

```python
plt.figure(figsize = (8,5))
plt.plot(inflation_long["Year"],inflation_long["Inflation"], marker = 'o')
plt.title("Inflation in Egypt in range(2014 : 2025)")
plt.ylabel("Inflation (%)")
plt.grid()
plt.show()
```

# Task 2.0: GDP per Capita Growth

```python
gdp_growth = pd.read_csv('GDP.csv', skiprows = 4)
gdp_growth.head()
```

# 2.1 Egypt GDP

```python
gdp_egypt_growth = gdp_growth[gdp_growth["Country Name"] == "Egypt, Arab Rep."]
gdp_egypt_growth.head()
```

```python
gdp_sel = gdp_egypt_growth[years].head()
gdp_sel
```

# 2.2 Reset Indexing

```python
gdp_long = gdp_sel.T.reset_index()
gdp_long.columns = ["Year","GDP_growth"]
gdp_long.head()
```

```python
gdp_long["Year"] = gdp_long["Year"].astype(int)
```

```python
gdp_long.info()
```

# 2.3 Ploting Result

```python
plt.figure(figsize = (8,5))
plt.plot(gdp_long["Year"],gdp_long["GDP_growth"], marker = "o")
plt.title("GDP_Growth_Per_Year")
plt.xlabel("Year")
plt.ylabel("GDP_Growth")
plt.grid()
plt.show()
```

# Task 3.0: Methodology 
## 3.1 Merging Data

```python
df = inflation_long.merge(gdp_long, on="Year", how="inner")
df.head()
```

## 3.2 GDP Calculations

```python
df["G"] = df["GDP_growth"] / 100
df["I"] = df["Inflation"] / 100
df["Real_Growth"] = ((1 + df["G"]) / (1 + df["I"])) - 1
df["RG_pct"] = df["Real_Growth"] * 100
df
```

```python
type(df)
```

```python
df = df.sort_values("Year").reset_index(drop = True)
df["PP"] = 100 * (df["Real_Growth"] + 1).cumprod() 
df[["Year", "RG_pct","PP"]]
```

# 4.0 Real GDP (Another Methodology)

```python
real_gdp = pd.read_csv("real_Gdp.csv", skiprows = 4)
real_gdp.head()
```

# 4.1 GDP Egypt

```python
gdp_egypt = real_gdp[real_gdp["Country Name"] == "Egypt, Arab Rep."]
years = [str(year) for year in range(2010, 2025)]
gdp_egypt_sel = gdp_egypt[years]
gdp_egypt_sel.head()
```

# 4.2 Reset indexing

```python
gdp_eg_long = gdp_egypt_sel.T.reset_index()
gdp_eg_long.columns = ["Year","GDP"]
gdp_eg_long["Year"] = gdp_eg_long["Year"].astype(int)
gdp_eg_long.head()
```

# 4.3 Refering to Base year 

```python
base = gdp_eg_long.loc[gdp_eg_long["Year"] == 2010, "GDP"].iloc[0]
gdp_eg_long["هndex_2010_100"] = 100 * (gdp_eg_long["GDP"] / base)
gdp_eg_long["هndex_2010_100"] = gdp_eg_long["هndex_2010_100"]/100
gdp_eg_long
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(8,5))

gdp_eg_long["Index_2010_1"] = gdp_eg_long["Index_2010_100"] / 100

sns.lineplot(
    data=gdp_eg_long,
    x="Year",
    y="Index_2010_1",
    marker="o",
    linewidth=2
)

plt.title("Year-over-Year Growth Rate", fontsize=13)
plt.ylabel("GDP Change of Rate")
plt.tight_layout()

plt.show()
```
